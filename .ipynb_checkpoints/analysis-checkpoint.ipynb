```python
%matplotlib inline
from IPython.core.pylabtools import figsize
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as stats
import pymc3 as pm
import theano.tensor as tt
```

```python
fake_data = stats.norm.rvs(loc=90,scale=20,size=5000)
figsize(12.5, 10)
plt.hist(fake_data, histtype='stepfilled', bins=30, alpha=0.85,
         label="OPT Processing Time, Days", color="#A60628", density=True)
plt.legend(loc="upper left")
plt.xlabel("Days")
```

```python
with pm.Model() as model:
    lambda_ = pm.Normal("lambda", mu = 80, sd = 20)
    sd = pm.Uniform("sd", lower=0,upper=45)
    processing_time = pm.Normal("Processing Time", mu=lambda_, sd=sd, observed=fake_data)
```

```python
with model:
    step = pm.Metropolis()
    trace = pm.sample(10000, tune=5000,step=step)
```

```python
lambda_samples = trace['lambda']
sigma_samples = trace['sd']
figsize(12.5, 10)
#histogram of the samples:
 
ax = plt.subplot(211)

plt.hist(lambda_samples, histtype='stepfilled', bins=30, alpha=0.85,
         label="posterior of $\lambda$", color="#A60628", density=True)
plt.legend(loc="upper left")
plt.title(r"""Posterior distributions of the variables
    $\lambda,\;\sigma$""")
plt.xlabel("$\lambda$ value")

ax = plt.subplot(212)
plt.hist(sigma_samples, histtype='stepfilled', bins=30, alpha=0.85,
         label="posterior of $\sigma$", color="#7A68A6", density=True)
plt.legend(loc="upper left")
plt.xlabel("$\sigma$ value");
```
